# Model creation

```python
x=lv.iloc[:,:10]
y=lv.target
```

```python
x
```

```python
y
```

```python

from collections import Counter# importing counter to check count of each label
from imblearn.over_sampling import SMOTE #for balancing the data
sm=SMOTE()#object creation
print(Counter(y))# checking count for each class 
x,y=sm.fit_resample(x,y)#applying sampling on target variable 
print(Counter(y))# checking count after sampling for  each class
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train
```

```python
# importing modules from sklearn libarary..

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC,LinearSVC
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron,SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,precision_score,recall_score,classification_report
from sklearn.model_selection import train_test_split
```

# LogisticRegression

```python
LR=LogisticRegression()
LR.fit(x_train,y_train) 
```

```python
lr_pr=LR.predict(x_test)
```

```python
lr_pr
```

```python
lr_acc=accuracy_score(y_test,lr_pr)# finding accuracy score 
lr_acc
```

```python
lr_f1=f1_score(y_test,lr_pr)
lr_f1
```

```python
cm=confusion_matrix(y_test,lr_pr)   
print(cm)
```

```python
score=classification_report(y_test,lr_pr)
print(score)
```
