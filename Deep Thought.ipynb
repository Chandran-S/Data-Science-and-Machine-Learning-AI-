# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#a=pd.read_excel()
xls = pd.ExcelFile('Behavioural Data Science data.xlsx')
performance = pd.read_excel(xls, 'Sheet 2 Performance scorecard', usecols=[1,4,5,6,7])
happiness = pd.read_excel(xls, 'Sheet 3 Happiness scorecard')
```

# Visualize with matplotlib and Seaborn

```python
plt.figure(figsize=(20,20), facecolor='white')#canvas size
plotnumber = 1

for column in week1:
    if plotnumber<=4 :
        ax = plt.subplot(2,3,plotnumber)
        sns.histplot(x=week1[column].dropna(axis=0)
                        ,hue=week1.Task)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Attrition',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
fig, ax = plt.subplots(3, 3)

ax[0][0].bar(week1.Task, week1.Neha,color='red')
ax[0][0].set_title('Neha',fontsize=10)
ax[0][0].tick_params(axis='x', rotation=40, labelsize=7)

ax[0][1].bar(week1.Task, week1.Amith,color='brown')
ax[0][1].set_title('Amith',fontsize=10)
ax[0][1].tick_params(axis='x', rotation=40, labelsize=7)

ax[0][2].bar(week1.Task, week1.Sachin,color='green')
ax[0][2].set_title('Sachin',fontsize=10)
ax[0][2].tick_params(axis='x', rotation=40, labelsize=7)

ax[1][0].bar(week2.Task, week2.Neha,color='red')
ax[1][0].set_title('Neha',fontsize=10)
ax[1][0].tick_params(axis='x', rotation=40, labelsize=7)

ax[1][1].bar(week2.Task, week2.Amith,color='brown')
ax[1][1].set_title('Amith',fontsize=10)
ax[1][1].tick_params(axis='x', rotation=40, labelsize=7)

ax[1][2].bar(week2.Task, week2.Sachin,color='green')
ax[1][2].set_title('Sachin',fontsize=10)
ax[1][2].tick_params(axis='x', rotation=40, labelsize=7)

ax[2][0].bar(week3.Task, week3.Neha,color='red')
ax[2][0].set_title('Neha',fontsize=10)
ax[2][0].tick_params(axis='x', rotation=40, labelsize=7)

ax[2][1].bar(week3.Task, week3.Amith,color='brown')
ax[2][1].set_title('Amith',fontsize=10)
ax[2][1].tick_params(axis='x', rotation=40, labelsize=7)

ax[2][2].bar(week3.Task, week3.Sachin,color='green')
ax[2][2].set_title('Sachin',fontsize=10)
ax[2][2].tick_params(axis='x', rotation=40, labelsize=7)

row_titles = ['WEEK-1', 'WEEK-2', 'WEEK-3']

for i in range(ax.shape[0]):
    ax[i][0].set_ylabel(row_titles[i], fontsize=12)

fig.subplots_adjust(wspace=0.5,hspace=0.6)
fig.set_figwidth(15)

for ax in fig.get_axes():
    ax.label_outer()
    
plt.show()
```
