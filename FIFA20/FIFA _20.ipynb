```python
def plt_top10 (dataframe,metric):
    met=fifa1[['short_name',metric]].sort_values(by=[metric],ascending=False)[:10]
    plt.title(f"Top 10 highest {metric} players in FIFA20")
    sns.barplot(y=met.short_name, x=met[metric])
    for index, value in enumerate(met[metric]):
        plt.text(value//2, index, str(round(value,2)))
plt_top10(fifa1, 'potential')

plt.tight_layout(pad=3.0)
```

```python
sns.countplot(data=fifa1,x='preferred_foot',palette='pink')
right,left=fifa1.preferred_foot.value_counts()
total=right+left
plt.text(0, left//2, left, fontsize = 20,  horizontalalignment='left')
plt.text(1, right//2, right, fontsize = 20, horizontalalignment='right')
```

```python
left_average=fifa1.groupby(['preferred_foot']).overall.mean()[0].round(2)
right_average=fifa1.groupby(['preferred_foot']).overall.mean()[1].round(2)
```

```python
print(f'Overall average of left foot players is {left_average} and ovrall average of right foot players is {right_average}')
```

```python
import scipy.stats
import math
```

```python
z_score = ((right-left)/total)/(math.sqrt(0.10*0.90*(1/left+1/right)))
p_value = scipy.stats.norm.sf(abs(z_score))*2
alpha=0.05
print('p_value is:',p_value)
if (p_value>alpha):
    print("Since p-value > α, we cannot reject the null hypthesis")
    print("There is no significant difference between left handed people and right handed people in sports")
else:
    print("Since p-value < α, we can reject the null hypthesis")
    print("There is a significant difference between left handed people and right handed people in sports")
```

```python
def nation_view(dataframe, country_name):
    return dataframe.loc[dataframe.nationality == country_name]
```
